```python
tf = pd.DataFrame(list(bag_of_words))
tf = tf.fillna(0)
tf
```

```python
tf_sparse = np.matrix(tf)
tf_sparse
```

```python
df = (tf > 0).sum(axis=0)
df
```

```python
idf = np.log(len(tf) / df)
idf
```

```python
tf_idf = tf * idf
tf_idf
```

```python
query = pd.Series(input())
```

```python
query = (query.str.lower().str.replace("[^\w\s]", " ").str.split()).apply(Counter)
```

```python
query = pd.DataFrame(list(query))
query
```

```python
a = query
B = tf_idf
dot = a.multiply(B).sum(axis=1)
dot
```

```python
# Calculate the denominators.
a_len = np.sqrt(a.multiply(a).sum())
b_len = np.sqrt(B.multiply(B).sum(axis=1))
print(a_len)
b_len
```

```python
# Calculate their ratio to obtain cosine similarities.
ratio = dot / (a_len * b_len)
ratio = ratio.fillna(0)
ratio
```

```python
cos_similarities = pd.DataFrame(ratio)[0]
most_similar = cos_similarities.sort_values(ascending=False)
most_similar
```
